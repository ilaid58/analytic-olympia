```python
#features that we will apply dummies :'GENDER', 'VEHICLE_YEAR'
def dummies(df, feature):
    df_dum = pd.get_dummies(df[feature]) # apply the dummies and the columns name the values of this feature
    df_dum.columns = feature + df_dum.columns # rename the column by the feature origin + df_dum columns 
    
    df[df_dum.columns] = df_dum # we create the columns in intial dataset with the df_dum
    df.drop(feature, axis=1, inplace=True) # drop the original columns in initial dataset
    
    return df

for i in ['GENDER', 'VEHICLE_YEAR']:
    df_train = dummies(df_train, i)
    df_test = dummies(df_test, i)
```

```python
# features that will apply ordinal encoder
feat_obj_ord_enc = ['AGE', 'DRIVING_EXPERIENCE', 'EDUCATION', 'INCOME', 'TYPE_OF_VEHICLE']
df_train[feat_obj_ord_enc] = OrdinalEncoder().fit_transform(df_train[feat_obj_ord_enc])
df_test[feat_obj_ord_enc] = OrdinalEncoder().fit_transform(df_test[feat_obj_ord_enc])
```

```python
feat_to_chang_type = [i for i in df_train.columns if df_train[i].dtypes=='float64' and i!='CREDIT_SCORE']
df_train[feat_to_chang_type] = df_train[feat_to_chang_type].astype('int64')
```

```python
# here we output the sigmoid function
features = [i for i in df_train.columns if i!='OUTCOME']
target = 'OUTCOME'
x = df_train[features].values
y = df_train[target].values
m,n = x.shape
lr = LinearRegression()
lr.fit(x, y)
w = lr.coef_
b = lr.intercept_
z = [np.dot(x[i], w) + b for i in range(m)]
f_wb = [1/(1+np.exp(-1*i)) for i in z]
plt.plot(z, f_wb)
plt.xlabel('Sigmoid function')
plt.show()
```

```python
# here we plot the relationship between each feature and target' using linearRegression()
lr = LinearRegression()
for i in features:
    x = df_train[[i]]
    y = df_train['OUTCOME']
    
    lr.fit(x, y)
    w = lr.coef_
    b = lr.intercept_
    f_wb = np.dot(x, w) + b
    
    plt.plot(x, f_wb, '.')
    plt.xlabel(i)
    plt.ylabel('OUTCOME')
    plt.show()
```

```python
p_values = []
p_coefs = []
for i in df_train.columns:
    p_coef, p_value = stats.pearsonr(df_train[i], df_train['OUTCOME'])
    
    p_values.append(p_value)
    p_coefs.append(p_coef)
relation = pd.DataFrame({'features':df_train.columns, 'p_values':p_values, 'p_coefs':p_coefs})
print(relation)
```

```python
df_best_features = relation[relation['p_values'] <0.78 ]
best_features = df_best_features['features']
best_features_test = [i for i in best_features.values]
best_features = [i for i in best_features.values]
best_features.remove('ID')
best_features_test.remove('OUTCOME')
```

```python
df_final = df_train[best_features]
df_final_test = df_test[best_features_test]
```

```python
# now after analyse and end up with the data, i will split the data into train, validation and test set
train_set = df_final[:10000]
valid_set = df_final[10000:12500]
test_set = df_final[12500:]
```

```python
best_features.remove('OUTCOME')
x_train, y_train = train_set[best_features], train_set[target]
x_valid, y_valid = valid_set[best_features], valid_set[target]
x_test, y_test = test_set[best_features], test_set[target]
```

```python
dt = DecisionTreeClassifier(max_depth=9, max_leaf_nodes=4)
dt.fit(x_train, y_train)
pred_valid = dt.predict(x_valid)
print(dt.score(x_valid, y_valid))
print(mse(pred_valid, y_valid))
```
